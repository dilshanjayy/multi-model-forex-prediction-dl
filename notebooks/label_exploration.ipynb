```python
import pandas as pd
import plotly.graph_objects as go

df = pd.read_csv("../data/raw_market/EURUSD_H1_20200101_20260131.csv", parse_dates=["time"])
df.set_index("time", inplace=True)
df.drop(columns=["spread", "real_volume"], inplace=True)

df
```

```python
month_df = df.loc["2026-01-01":"2026-01-31"]

fig = go.Figure(data=[go.Candlestick(x=month_df.index,
                open=month_df["open"],
                high=month_df["high"],
                low=month_df["low"],
                close=month_df["close"],
                name="Market Data")]
                )

fig.update_layout(
    title="2026 January EUR/USD H1 Price Action",
    yaxis_title="Price",
    xaxis_title="Time",
    template="plotly_dark",
    xaxis_rangeslider_visible=False,
    height=600
)

# identify missing timestamps in the hourly data
full_timeline = pd.date_range(start=month_df.index.min(), end=month_df.index.max(), freq="h")
missing_timestamps = full_timeline.difference(pd.DatetimeIndex(month_df.index))

# hide missing timestamps on the x-axis to avoid gaps in the candlestick chart
fig.update_xaxes(
    rangebreaks=[
        dict(values=missing_timestamps, dvalue=3600000)
    ]
)

fig.show()
```
